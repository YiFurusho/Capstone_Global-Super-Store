```python

# Step 1: Load the dataset
path='/content/Capstone_python_Global_Superstore2.xlsx'
pd.read_excel(path)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Check the columns to identify the ones you need
df = pd.read_excel(path)
print(df.columns)  # To see the column names and their structure
# Step 2: Group data by 'Region' and sum the 'Sales'
region_sales = df.groupby('Region')['Sales'].sum()
# Step 4: Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(region_sales.index, region_sales.values, color='skyblue')
plt.title('Total Sales by Region', fontsize=12)
plt.xlabel('Region', fontsize=10)
plt.ylabel('Total Sales', fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)

plt.show()
```

```python
## 2. How does profit vary by category?
# Group by Category and calculate average Profit
category_profit = df.groupby('Category')['Profit'].mean()

# Plot
plt.figure(figsize=(8, 5))

plt.bar(category_profit.index, category_profit.values, color='lightgreen')

plt.title('Average Profit by Category')
plt.xlabel('Category')
plt.ylabel('Average Profit')

plt.show()
```

```python
## 3. What is the distribution of order priorities?
# Count the occurrences of each Order Priority
order_priority_counts = df['Order Priority'].value_counts()

# Plot
plt.figure(figsize=(6, 6))

plt.pie(order_priority_counts, labels=order_priority_counts.index, autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightblue', 'lightgreen'])

plt.title('Distribution of Order Priorities')

plt.show()
```

```python
## 4. How does sales vary over time?
# Convert Order Date to datetime and group by month
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

# Plot
plt.figure(figsize=(24, 10))

plt.plot(monthly_sales.index.astype(str), monthly_sales.values, marker='o', color='blue')

plt.title('Monthly Sales Over Time', fontsize = 24)
plt.xlabel('Month', fontsize = 18)
plt.ylabel('Total Sales ($)', fontsize = 18)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)

plt.show()
```

```python
## 5. What is the relationship between sales and profit?
# Plot
plt.figure(figsize=(8, 6))

plt.scatter(df['Sales'], df['Profit'], alpha=0.1, color='purple')

plt.title('Relationship Between Sales and Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')

plt.show()
```

```python
## 6. Which cities have the highest sales?
# Group by City and sum Sales, then select top 10
city_sales = df.groupby('City')['Sales'].sum().nlargest(10)

plt.figure(figsize=(10, 6))
plt.barh(city_sales.index, city_sales.values, color='orange')

plt.title('Top 10 Cities by Sales')
plt.xlabel('Total Sales ($)')
plt.ylabel('City')

plt.show()
```

```python
## 7. How does discount affect profit?
# Plot
plt.figure(figsize=(8, 6))

plt.scatter(df['Discount'], df['Profit'], alpha=0.5, color='red')

plt.title('Effect of Discount on Profit')
plt.xlabel('Discount')
plt.ylabel('Profit')

plt.show()
```

```python
## 8. What is the distribution of sales across different ship modes?
# Group by Ship Mode and sum Sales
ship_mode_sales = df.groupby('Ship Mode')['Sales'].sum()

# Plot
plt.figure(figsize=(8, 5))

plt.bar(ship_mode_sales.index, ship_mode_sales.values, color='teal')

plt.title('Sales Distribution by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Total Sales')

plt.show()
```

```python
## 9. How does quantity ordered vary by sub-category?
# Group by Sub-Category and sum Quantity
subcategory_quantity = df.groupby('Sub-Category')['Quantity'].sum()

# Plot
plt.figure(figsize=(12, 6))

plt.bar(subcategory_quantity.index, subcategory_quantity.values, color='skyblue')

plt.title('Total Quantity Ordered by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Quantity')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)

plt.show()
```

```python
## 10. What is the distribution of profits across different markets?
# Plot
plt.figure(figsize=(10, 6))

df.boxplot(column='Profit', by='Market', grid=False, patch_artist=True, boxprops=dict(facecolor='gold'))

plt.title('Profit Distribution by Market')
plt.suptitle('')  # Remove automatic title
plt.xlabel('Market')
plt.ylabel('Profit')

plt.show()
```
